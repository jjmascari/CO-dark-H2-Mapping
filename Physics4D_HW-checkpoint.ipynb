```python
import numpy as np
import matplotlib.pyplot as plt

# Define Gaussian function
def gaussian(x, A, mu, sigma):
    return A * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))

# Parameters
A = 1       # Amplitude
mu = 0      # Mean (center of the peak)
sigma = 1  # Standard deviation (spread)

# Generate x values
x = np.linspace(-5, 5, 400)  # Range from -5 to 5 with 400 points

# Compute Gaussian values
y = gaussian(x, A, mu, sigma)

# Plot the function
plt.plot(x, y, label=f'Gaussian: μ={mu}, σ={sigma}', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gaussian Function')
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Simulation parameters
wavelength = 1.0  # Wavelength of light (arbitrary units)
slit_widths = [2.0, 1.0, 0.5, 0.2]  # Different slit widths to compare
screen_distance = 10  # Distance to observation screen
x_range = np.linspace(-5, 5, 500)  # Screen position values

# Function to compute far-field diffraction pattern
def diffraction_pattern(x, d, wavelength):
    if d < wavelength:
        # Extreme diffraction case: Near-isotropic spreading
        return np.ones_like(x)  # Uniform intensity
    else:
        k = 2 * np.pi / wavelength  # Wave number
        beta = (k * d * x) / (2 * screen_distance)
        intensity = (np.sinc(beta / np.pi))**2  # Single-slit diffraction formula
        return intensity

# Plot results
plt.figure(figsize=(8, 6))
colors = cm.viridis(np.linspace(0.2, 1, len(slit_widths)))

for i, d in enumerate(slit_widths):
    intensity = diffraction_pattern(x_range, d, wavelength)
    plt.plot(x_range, intensity, label=f'Slit Width = {d}λ', color=colors[i])

plt.xlabel("Screen Position")
plt.ylabel("Intensity")
plt.title("Diffraction Patterns for Different Slit Widths")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal

# Define parameters
slit_width = 1.0  # Width of each slit
distance_between_slits = 3.0  # Center-to-center distance between the two slits
x = np.linspace(-5, 5, 1000)  # Spatial domain

# Single slit function (rectangular function)
def single_slit(x, width):
    return np.where(np.abs(x) <= width / 2, 1, 0)

# Dirac comb function (two delta functions for double slit)
def dirac_comb(x, separation):
    return np.where(np.isclose(x, -separation / 2, atol=0.05), 1, 0) + np.where(np.isclose(x, separation / 2, atol=0.05), 1, 0)

# Compute single slit and Dirac comb
single_slit_function = single_slit(x, slit_width)
dirac_comb_function = dirac_comb(x, distance_between_slits)

# Convolution: double slit function
conv_result = scipy.signal.convolve(single_slit_function, dirac_comb_function, mode='same')

# Compute Fourier Transform (diffraction pattern)
k = np.fft.fftfreq(len(x), d=(x[1] - x[0]))  # Frequency domain
fourier_transform = np.abs(np.fft.fft(conv_result))
fourier_transform = np.fft.fftshift(fourier_transform)  # Shift for visualization

# Plot results
fig, axs = plt.subplots(4, 1, figsize=(8, 10))

axs[0].plot(x, single_slit_function, label='Single Slit Function', color='blue')
axs[0].set_title("Single Slit Aperture Function")
axs[0].legend()

axs[1].plot(x, dirac_comb_function, label='Dirac Comb (Two Delta Functions)', color='red')
axs[1].set_title("Slit Positions (Dirac Comb)")
axs[1].legend()

axs[2].plot(x, conv_result, label='Double Slit Function (Convolution)', color='green')
axs[2].set_title("Double Slit Function (Single Slit * Dirac Comb)")
axs[2].legend()

axs[3].plot(k, fourier_transform, label='Diffraction Pattern', color='purple')
axs[3].set_title("Fourier Transform (Diffraction Pattern)")
axs[3].set_xlim([-10, 10])
axs[3].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Gaussian function
A = 1         # Amplitude
mu = 0        # Mean (center)
sigma = 0.0001     # Standard deviation

# Generate x values
x = np.linspace(-5, 5, 1000)

# 1D Gaussian function
G_x = A * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Fourier transform of the 1D Gaussian
G_k = np.fft.fftshift(np.fft.fft(G_x))
k = np.fft.fftshift(np.fft.fftfreq(len(x), d=(x[1] - x[0])))

# Squared magnitude of the Fourier transform
G_k_squared = np.abs(G_k)**2

# Plot the 1D Gaussian
plt.figure(figsize=(8, 6))
plt.plot(x, G_x, label='1D Gaussian')
plt.xlabel('x')
plt.ylabel('G(x)')
plt.title('1D Gaussian Function')
plt.legend()
plt.grid(True)
plt.show()

# Plot the squared Fourier transform
plt.figure(figsize=(8, 6))
plt.plot(k, G_k_squared, label='Squared Fourier Transform')
plt.xlabel('k')
plt.ylabel('|G(k)|^2')
plt.title('Squared Fourier Transform of 1D Gaussian')
plt.legend()
plt.grid(True)
plt.show()

# Parameters for 2D Gaussian function
mu_x, mu_y = 0, 0
sigma_x, sigma_y = 1, 1

# Generate x and y values
grid_size = 100
x = np.linspace(-5, 5, grid_size)
y = np.linspace(-5, 5, grid_size)
X, Y = np.meshgrid(x, y)

# 2D Gaussian function
G_xy = A * np.exp(-((X - mu_x)**2 / (2 * sigma_x**2) + (Y - mu_y)**2 / (2 * sigma_y**2)))

# Fourier transform of the 2D Gaussian
G_kx_ky = np.fft.fftshift(np.fft.fft2(G_xy))
kx = np.fft.fftshift(np.fft.fftfreq(grid_size, d=(x[1] - x[0])))
ky = np.fft.fftshift(np.fft.fftfreq(grid_size, d=(y[1] - y[0])))

# Squared magnitude of the 2D Fourier transform
G_kx_ky_squared = np.abs(G_kx_ky)**2

# Plot the 2D Gaussian
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, G_xy, cmap='viridis')
plt.colorbar(label='G(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Gaussian Function')
plt.show()

# Plot the squared 2D Fourier transform
plt.figure(figsize=(8, 6))
plt.contourf(kx, ky, G_kx_ky_squared, cmap='viridis')
plt.colorbar(label='|G(kx, ky)|^2')
plt.xlabel('kx')
plt.ylabel('ky')
plt.title('Squared Fourier Transform of 2D Gaussian')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the concentric rings aperture
grid_size = 500
num_rings = 5
ring_spacing = 2
ring_width = 2

# Create a 2D grid
x = np.linspace(-100, 100, grid_size)
y = np.linspace(-100, 100, grid_size)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)

# Define the concentric rings aperture function
aperture = np.zeros_like(R)
for i in range(num_rings):
    ring_radius = (i + 1) * ring_spacing
    aperture += np.logical_and(R >= ring_radius - ring_width / 2, R <= ring_radius + ring_width / 2)

# Compute the Fourier transform of the aperture
ft_aperture = np.fft.fftshift(np.fft.fft2(aperture))
intensity_pattern = np.abs(ft_aperture)**2

# Normalize the intensity pattern for better visualization
intensity_pattern /= np.max(intensity_pattern)

# Plot the aperture function
plt.figure(figsize=(8, 6))
plt.imshow(aperture, cmap='gray', extent=[-100, 100, -100, 100])
plt.colorbar(label='Aperture')
plt.title('Concentric Rings Aperture Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot the interference (diffraction) pattern
plt.figure(figsize=(8, 6))
plt.imshow(intensity_pattern, cmap='inferno', extent=[-100, 100, -100, 100])
plt.colorbar(label='Intensity')
plt.title('Interference Pattern of Concentric Rings Aperture')
plt.xlabel('kx')
plt.ylabel('ky')
plt.show()
```
